```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Model 1

```python
df_1 = pd.read_csv('data/model_loss/mae_model_1.csv')
plt.plot(df_1['epoch'], df_1['loss'])
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.xticks(df_1['epoch'],)
plt.title("MSE of Model 1 Training")
plt.show()
```

# Model 2

```python
plt.plot(df_1['epoch'], df_1['loss'], label='model_1')
plt.plot(df_2['epoch'], df_2['loss'], label='model_2')
plt.plot(df_3['epoch'], df_3['loss'], label='model_3')

plt.title("MSE of MAE Models Training")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.xticks(range(0,14))
plt.legend()
```
